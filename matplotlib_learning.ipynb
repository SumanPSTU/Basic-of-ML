```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [10,13,6,40,33]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.title("Matplot lib learning")
plt.xlabel("X level")
plt.ylabel("Y level")
plt.show()
```

```python
x = [1,2,3,4]

y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x,y1,label="Sales")
plt.plot(x,y2,label="Profit")

plt.legend()

plt.show()
```
