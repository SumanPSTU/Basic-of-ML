**#indexing and slicing (Critical for ML)**

```python
a = np.array([1,2,3,4,5,6,7,8])
print(a)
print(a[0])
print(a[2:5])
print(a[1:4])
```

**#2D slicing**

```python
b = np.array([[1,2,3,4],[5,6,7,8]])
print(b[0,1])
print(b[0,3])
print(b[:,1])
print(b[1,:])
```

**#Mathemetical Operation**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

**#Broadcasting (SUPER IMPORTANT)**

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([1,2,3])
print(a+b)
```

**#matrix multiplication***

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(np.dot(a,b))

print(np.linalg.det(a))
```

```python
print(np.linalg.inv(a))
print(np.linalg.eig(a))
```

```python
a = np.array([1,2,3,4,5])
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.var(a))
print(np.min(a))
print(np.max(a))
print(np.sum(a))
```

**#For axis***

```python
b = np.array([[1,2,3,4],[6,7,8,9]])
np.mean(b,axis=0)
np.mean(b,axis=1)
```
